# Test FNPCA

Performs FNPCA on cached photometry data

```python
from erebus.photometry_data import PhotometryData
from erebus.wrapped_fits import WrappedFits
from erebus.utility import fits_file_utils as f_utils
from erebus.frame_normalized_pca import perform_fnpca_on_full_frame
import matplotlib.pyplot as plt
```

```python
folder = "./mast_lhs1478b"
visits = f_utils.get_fits_files_visits_in_folder(folder)
fits = WrappedFits(folder, visits[0])
photometry = PhotometryData(fits, 5, (12,20))
eigenvalues, eigenvectors, _ = perform_fnpca_on_full_frame(fits.frames[20:-40], photometry.radius, photometry.annulus_start, photometry.annulus_end)
plt.plot(eigenvalues[0])
plt.show()
plt.imshow(eigenvectors[0])
plt.show()
```
